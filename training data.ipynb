```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# digits to delete
digits_to_delete = [1, 3,5,7]


indices_to_delete = np.isin(y_train, digits_to_delete)
x_train_filtered = x_train[~indices_to_delete]
y_train_filtered = y_train[~indices_to_delete]

# Preprocess the data
x_train_filtered = x_train_filtered.reshape((-1, 28, 28, 1)).astype('float32') / 255.0
y_train_filtered = to_categorical(y_train_filtered, 10)  # One-hot encode the labels

# Verify the new dataset shape
print("Original training data shape:", x_train.shape)
print("Filtered training data shape:", x_train_filtered.shape)

```
